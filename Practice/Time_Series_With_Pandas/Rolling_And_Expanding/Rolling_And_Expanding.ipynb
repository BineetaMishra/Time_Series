```python
import pandas as pd
```

```python
df= pd.read_csv('starbucks.csv',index_col='Date',parse_dates=True)
```

```python
df['Close'].plot(figsize=(12,5))
```

## Rolling Mean

```python
# take average of evry 7 days.

df.rolling(window=7,).mean()
```

### Here first 7 values are empty because there is not much to do with that data. The average is found from the next 7

```python
df1= df.rolling(window=7,).mean()
```
