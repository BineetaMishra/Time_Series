```python
import numpy as np
import pandas as pd
```

```python
airline = pd.read_csv('airline_passengers.csv',index_col='Month',parse_dates=True)
```

```python
airline.head()
```

## Dropping the Missing Components

```python
airline= airline.dropna()
```

```python
airline.head()
```

```python
airline.plot()
```

## Multiplicative Model
### As the trend is increasing slightly more than the linear we use Multiplicative Model.

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result= seasonal_decompose(airline['Thousands of Passengers'],model='multiplicative')
```

# to print the entire rows in dataset use:
pd.options.display.max_rows = 999

# Components

## Trend Components

```python
pd.options.display.max_rows = 999
print(result.trend)
```

## Seasonal Components

```python
result.seasonal
```

## Residual / Error Components

```python
result.resid
```

```python
result.plot();
```
