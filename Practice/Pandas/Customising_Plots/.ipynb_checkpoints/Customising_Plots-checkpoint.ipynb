### Legend, Y axis, X axis, Title

```python
title= "MY PLOT TITLE"
xlabel= 'My X Data'
ylabel= 'My Y Data'
```

```python
ax= df2['c'].plot.line(figsize=(20,3),ls='-.',c='red',lw=8)
ax.set(xlabel=xlabel,ylabel=ylabel)
```

```python
ax= df2['c'].plot.line(figsize=(20,3),ls='-.',c='red',lw=8,title=title)
ax.set(xlabel=xlabel,ylabel=ylabel)
```
